```python
def twoPtForwardDiff(x,y):
    """function takes the dy/dx of element ahead of current point
    Function takes the backwards difference at the last poin"""
    
    dydx = np.zeros(len(x))
    
    dydx[0:-1] = (y[1:] - y[:-1]) / (x[1:] - x[:-1])
    
    dydx[-1] = (y[-1] - y[-2]) / (x[-1] - x[-2])
    
    return dydx

def twoPtCenteredDiff(x,y):
    """function takes the slope of the line between the two
    points above and below it. the first and last points
    are calculated using the regular forwards and backward
    derivatives though"""
    
    dydx = np.zeros(len(x))
    
    dydx[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2])
    
    dydx[0] = (y[1] - y[0])/(x[1] - x[0])
    
    dydx[-1] = (y[-1] - y[-2]) / (x[-1] - x[-2])

    return dydx
```

```python
x = np.arange(1,10,.1)
y = x**3

dydx = twoPtForwardDiff(x,y)
dydx2 = twoPtCenteredDiff(x,y)

plt.plot(x,dydx)
plt.plot(x,dydx2)
plt.show()
```

In order to calculate values for dydx[0], dydx[1], dydx[-1] and dydx[-2] you must use the lowest order finite difference method (like the ones you defined in your twoPtForwardDiff(x,y) and twoPtCenteredDiff(x,y) functions).
```python
def fourPtCenteredDiff(x,y):
    
    difference = max(np.diff(x))
    
    dydx = np.zeros(len(x))
    
    dydx[1] = (y[2] - y[0])/(x[2] - x[0])
    
    dydx[0] = (y[1] - y[0])/(x[1] - x[0])
    
    dydx[-1] = (y[-1] - y[-2]) / (x[-1] - x[-2])
    
    dydx[-2] = (y[-1] - y[-3])/(x[-1] - x[-3])
    
    dydx[2:-2] = (y[:-4] - 8*y[1:-3] + 8*y[3:-1]
                  - y[4:])/(12*difference)
    
    return dydx
```
