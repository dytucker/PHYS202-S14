(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
from scipy.integrate import odeint
```

```python
def deriv(theta_current,t):
    g = 9.8
    ell = 0.5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([dtheta,domega])

t = np.arange(0.,20.+0.01,0.01)


th0_1 = np.pi/6.
th0_2 = np.pi/4.
th0_3 = np.pi/2.
th0_4 = 3*np.pi/4.
th0_5 = np.pi-.2


omega0 = 0
theta_int1=np.array([th0_1,omega0])
theta_int2=np.array([th0_2,omega0])
theta_int3=np.array([th0_3,omega0])
theta_int4=np.array([th0_4,omega0])
theta_int5=np.array([th0_5,omega0])

theta1 = odeint(deriv,theta_int1,t)
theta2 = odeint(deriv,theta_int2,t)
theta3 = odeint(deriv,theta_int3,t)
theta4 = odeint(deriv,theta_int4,t)
theta5 = odeint(deriv,theta_int5,t)
```

```python
plt.plot(t,theta1[:,0],'r--',label="theta$_0 = \pi/6$")
plt.plot(t,theta2[:,0],'g',label="theta$_0 = \pi/4$")
plt.plot(t,theta3[:,0],'b--',label="theta$_0 = \pi/2$")
plt.plot(t,theta4[:,0],'gray',label="theta$_0 = 3\pi/4$")
plt.plot(t,theta5[:,0],'purple',label="theta$_0 = \pi/-.2$")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
#plt.ylim(-1,1.5)
plt.xlim(0,20)
plt.grid()
plt.legend()
plt.show()
```

As $\theta_0$ approaches $\pi$, the pendulum takes longer to complete a revolution and reaches higher peaks.

(a) Use Euler's method to solve the following first-order differential equation with given initial condition $y(0) = 0$.
$dy/dx=x+2y$
Find the value of y when x = 1, that is y(1), using a first a step size of 0.25, and then 0.02.

```python
def EulerSolver_1():
    xmax = 1.25 # maximum time
    dx = 0.25 # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    for i in range(len(x)-1):
            x[i+1] = x[i] + dx
            y[i+1] = y[i] + dx * (x[i]+2*y[i])
            
    return x,y
x1,y1 = EulerSolver_1()
```

```python
plt.plot(x1,y1,label="y vs. x for time step = 0.25")
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend(loc = "upper left")
plt.show()
```

```python
def EulerSolver_2():
    xmax = 1 # maximum time
    dx = 0.02 # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    for i in range(len(x)-1):
            x[i+1] = x[i] + dx
            y[i+1] = y[i] + dx * (x[i]+2*y[i])
            
    return x,y
x2,y2 = EulerSolver_2()
```

```python
plt.plot(x2,y2,label="y vs. x for time step = 0.02")
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend(loc = "upper left")
plt.show()
```

According to the plots, y(1) is about 5.5 when the time step is 0.25 and almost 1 when the time step is 0.02.

(c) Now solve the same problem using Euler-Cromer and the Midpoint method, using the same pair of step sizes and plot the results along with the exact solution. Instead of plotting the solutions, plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
def Euler_CromerSolver1():
    xmax = 1 # maximum time
    dx = 0.25 # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    for i in range(len(x)-1):
            x[i+1] = x[i] + dx
            y[i+1] = y[i] + dx * (x[i+1]+2*y[i])
            
    return x,y
x1,y1 = Euler_CromerSolver1()

def Euler_CromerSolver2():
    xmax = 1 # maximum time
    dx = 0.25 # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    for i in range(len(x)-1):
            x[i+1] = x[i] + dx
            y[i+1] = y[i] + dx * (x[i+1]+2*y[i])
            
    return x,y
x2,y2 = Euler_CromerSolver2()

def MidpointSolver_1():
    xmax = 1 # maximum time
    dx = 0.25 # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    for i in range(len(x)-1):
            x[i+1] = x[i] + dx
            y[i+1] = y[i] + dx * 0.5 * ((x[i+1] + 2*y[i]) + (x[i+1]+2*y[i]))
            
    return x,y
x1m,y1m = MidpointSolver_1()

def MidpointSolver_2():
    xmax = 1 # maximum time
    dx = 0.02 # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    for i in range(len(x)-1):
            x[i+1] = x[i] + dx
            y[i+1] = y[i] + dx * 0.5 * ((x[i+1] + 2*y[i]) + (x[i+1]+2*y[i]))
            
    return x,y
x2m,y2m = MidpointSolver_2()
```

```python
t = np.arange(0,1,.001)
exact = 0.25*np.exp(2*t)-0.5*t-0.25

plt.plot(x1,y1,label="Euler-Cromer for time step = 0.25")
plt.plot(t,exact, label = "Analytic solution")
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend(loc = "upper left")
plt.show()

plt.plot(x2,y2,label="Euler-Cromer for time step = 0.02")
plt.plot(t,exact, label = "Analytic solution")
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend(loc = "upper left")
plt.show()

plt.plot(x1m,y1m,label="Midpoint method for time step = 0.25")
plt.plot(t,exact, label = "Analytic solution")
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend(loc = "upper left")
plt.show()

plt.plot(x2m,y2m,label="Midpoint Method for time step = 0.02")
plt.plot(t,exact, label = "Analytic solution")
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
plt.legend(loc = "upper left")
plt.show()
```

The best solution is obtained by the midpoint method with time step of 0.02.

We can use numerical solvers and widgets to explore the dynamics of a projectile that is subject to a drag force proportional to the square of its velocity - "quadratic drag".
The force on the projectile is given by
∑F⃗ =mg⃗ −cv2v^=mg⃗ −cvv⃗ =mg⃗ −cv2x+v2y−−−−−−√v⃗ =ma⃗ 
Where the second term is the drag force, in the direction opposite that of the velocity vector. Re-writing these in terms of the first-order differential equations for vx and vy gives
mv˙x=−cv2x+v2y−−−−−−√vx
mv˙y=mg−cv2x+v2y−−−−−−√vy
taking the positive y direction to be down. Notice that these are coupled equations. There are both vx and vy terms in each one. The drag coefficient c depends on the shape and density of the projectile. We can re-write this system of equations in terms of the terminal velocity - the point at which the gravitational force balances the drag force. This happens when mg=cv2. Let's denote the terminal velocity as vter=mg/c−−−−−√ and re-write our equations in terms of vter.
v˙x=−gvxv2terv2x+v2y−−−−−−√
v˙y=−g(1+vyv2terv2x+v2y−−−−−−√)

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far the projectile can travel in the x direction before landing.

```python
def Projectile_deriv(v,t):
    
    c = 1
    g = 9.81
    m = 1
    v_terr = 44
    
    x_0 = v[0]
    y_0 = v[1]
    xx = v[2]
    yy = v[3]
    
    dvdt_x = -g*(x_0/(v_terr**2))*np.sqrt(x_0**2+y_0**2)
    dvdt_y = -g*(1+(y_0/(v_terr**2))*np.sqrt(x_0**2+y_0**2))
    
    return np.array([dvdt_x,dvdt_y,x_0,y_0])
```

```python
t = np.arange(0,200)

vx_0 = 19.28
vy_0 = 23

result = odeint(Projectile_deriv,np.array([vx_0,vy_0,0,0]),t)
```

```python
vx = result[:,0]
vy = result[:,1]

plt.plot(t,vx,label="x velocity")
plt.plot(t,vy,label = "y velocity")
plt.xlabel("time (s)")
plt.ylabel("velocity (m/s)")
plt.legend(loc='best')
plt.show()
```

(b) Now create an interact object that uses sliders for the initial speed, terminal velocity, launch angle, and time. Have it plot two curves - one representing the solution with quadratic drag and the other in the absence of drag:

```python
from IPython.html.widgets import interact, interactive
```

```python
def Projectile_deriv(v,t):
    
    c = 1
    g = 9.81
    m = 1
    v_terr = v[2]
    
    x_0 = v[0]
    y_0 = v[1]
    xx = v[3]
    yy = v[4]
    
    dvdt_x = -g*(x_0/(v_terr**2))*np.sqrt(x_0**2+y_0**2)
    dvdt_y = -g*(1+(y_0/(v_terr**2))*np.sqrt(x_0**2+y_0**2))
    
    return np.array([dvdt_x,dvdt_y,x_0,y_0])

def simple_position(v,time):
    v_x = v[0]
    v_y = v[1]
    dt = 0.01
    
    x=[0]
    #change this part to alder y_0
    y=[1.5]

    
    for i in range(1,len(time)):
        x.append(v_x*i*dt)
        y.append(v_y*i*dt - 9.81/2.*(i*dt)**2 + 1.5)
        if i > 2 and y[i] <= 0:
            x = x[0:i]
            y = y[0:i]
            break
            
    return x,y

def Projectile_widget(speed=30,launch_angle=50,terminal_velocity=44,time=10):

    dt = .01
    t = np.arange(0,time,dt)

    vx_0 = speed*np.cos(launch_angle*np.pi/180.)
    vy_0 = speed*np.sin(launch_angle*np.pi/180.)
    v_terr = terminal_velocity
    
    x_0 = 0
    y_0 = 1.5

    result1 = odeint(Projectile_deriv,np.array([vx_0,vy_0,v_terr,x_0,y_0]),t)
    
    x_simp,y_simp = simple_position(np.array([vx_0,vy_0]),t)
    

    vx = result1[:,0]
    vy = result1[:,1]

    x = np.zeros(len(vx))
    y = np.zeros(len(vy))

    for i in range(1,len(vx)):
        x[i] = vx[i]*dt + x[i-1]
        y[i] = vy[i]*dt + y[i-1]
        if i > 2 and y[i] <= 0:
            x = x[0:i]
            y = y[0:i]
            break
        
    plt.plot(x,y,label="Dragged projectile path")
    plt.plot(x_simp,y_simp,label="No drag projectile path")
    #plt.plot(x,np.zeros(len(x)),label="ground")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(loc='best')
    plt.ylim = (0,max(y)+3)
    plt.show()
    
v = interact(Projectile_widget,speed=30,launch_angle=50,terminal_velocity=44,time=10)
```
