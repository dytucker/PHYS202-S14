1 . Download the linked file called sunspots.txt, which contains the observed number of sunspots on the Sun for each month since January 1749. The file contains two columns of numbers, the first being the month and the second being the sunspot number. (See note below∗ for how to download it.)

(a) Write some code that reads in the data and makes a graph of sunspots as a function of time.
```python
month, spots = np.loadtxt('sunspots.txt', unpack=True, usecols=[0,1])
#plt.plot(month,spots)

#turning the months into years after 1750
month /=12
month += 1750

#setting line width & color
plt.plot(month, spots, color="peru", linewidth=.5, linestyle="-", label="number of sunspots")
plt.legend(loc='upper center')

# set x,y limits
plt.xlim(month.min(), month.max())
plt.ylim(spots.min() * 1.1, 300)

#setting tick marks
x_ticks = range(1750,2025,25)
plt.xticks(x_ticks)
plt.yticks(range(0,301,25))

#making labels
plt.xlabel("Year", color = "navy", fontsize = 20)
plt.ylabel("Number of sunspots", color = "SlateGray", fontsize = 20)

plt.show()
```

(b) Now add code to calculate the running average of the data. Plot both the original data and the running average on the same graph, only over the range covered by the first 1000 data points.

```python
def calc_average(values):
    total_sum = 0
    so_far = 0
    average = []
    for i in values:
        so_far += 1.0
        total_sum += float(i)
        if so_far == 1001:
            break
        average.append(total_sum / so_far)
    return average
```

```python
running_total = (calc_average(spots))
print len(running_total)
```

```python
month, spots = np.loadtxt('sunspots.txt', unpack=True, usecols=[0,1])
#plt.plot(month,spots)

#turning the months into years after 1750
month /=12
month += 1750

#setting line width & color
plt.plot(month, spots, color="peru", linewidth=.5, linestyle="-", label="number of sunspots")
plt.plot(month[0:1000:], running_total, color="purple", linewidth=2, linestyle="-", label="Running Average of Spots")
plt.legend(loc='upper center')
plt.plot()

# set x,y limits
plt.xlim(month.min(), month.max())
plt.ylim(spots.min() * 1.1, 300)

#setting tick marks
x_ticks = range(1750,2025,25)
plt.xticks(x_ticks)
plt.yticks(range(0,301,25))

#making labels
plt.xlabel("Year", color = "navy", fontsize = 20)
plt.ylabel("Number of sunspots", color = "SlateGray", fontsize = 20)

plt.show()
```

5 . Determine the electric potential for a quadrupole consisting of

$q1 = -1.0 nC$ at $(-2 m, 0 m)$

$q2 = -1.0 nC$ at $(+2 m, 0 m)$

$q3 = +1.0 nC$ at $(0 m, -2 m)$

$q4 = +1.0 nC$ at $(0 m, +2 m)$

(a) Make a surface plot on the range $-4 <(x,y)< 4$ using 101 bins in x and 101 bins in y. In this case, do not let $k = q = d = 1$. Use dimensionful quantities and compute the value of the potential in volts.
Recall $k = 8.9875517873681764 × 109 N⋅ m2/C2$.

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

#make the x-y grid to cover the plane -4<x<4 and -4<y<4 in steps of 0.08 (or 101 bins)
dx = 0.08
dy = 0.08
#redefining limits
x = np.arange(-4, 4, dx)
y = np.arange(-4, 4, dy)
x, y = np.meshgrid(x, y)

#make the function, V(x,y)
k = 8.9875517873681764 * 109; 
q1 = -1.0;
q2 = -1.0;
q3 = 1.0;
q4 = 1.0;
d = 2
V=(k*q3/(x**2+(y-d)**2)**(1/2.)) + (k*q4/(x**2+(y+d)**2)**(1/2.)) - (k*q1/((x+d)**2+(y)**2)**(1/2.)) - (k*q2/((x-d)**2+(y)**2)**(1/2.))

#To use 3D graphics in matplotlib, we first need to create an axes instance of the class Axes3D. 
#3D axes can be added to a matplotlib figure canvas in exactly the same way as 2D axes, but a 
#convenient way to create a 3D axis instance is to use the projection='3d' keyword argument to 
#the add_axes or add_subplot functions.

fig = figure()

#Call the "get current axis" method on the figure object and set the projection to 3D
ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, V, rstride=1, cstride=1, linewidth=0,cmap=cm.coolwarm)

#To understand what the arguments to the plotting function do, try calling the help: 
#help(Axes3D.plot surface). In particular, investigate the colormap option. You can 
#also add a color bar to represent the value of the potential at each space point (x, y)
cb = fig.colorbar(p, shrink=0.5)

xlabel("x")
ylabel("y")

#show()
savefig("Quadpole.png")
```

(b) Now make a contour plot of the same data, with appropriate contour levels to see the shape of the potential.

```python
levels = np.array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
levels *= 2000
plt.contour(x,y,V,levels,linewidths=4,cmap=cm.coolwarm);

plt.show()
```
