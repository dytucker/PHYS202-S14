(b) Now add code to calculate the running average of the data. Plot both the original data and the running average on the same graph, only over the range covered by the first 1000 data points.

```python
def calc_average(values):
    total_sum = 0
    so_far = 0
    average = []
    for i in values:
        so_far += 1.0
        total_sum += float(i)
        if so_far == 1001:
            break
        average.append(total_sum / so_far)
    return average
```

```python
running_total = (calc_average(spots))
print len(running_total)
```
